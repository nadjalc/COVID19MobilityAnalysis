```python
import pandas as pd
import glob
```

```python
pattern = './week_data/*.csv'
csv_files = glob.glob(pattern)

frames = []

for csv in csv_files:
    df = pd.read_csv(csv, sep=";", index_col='timestamp', parse_dates=True)
    frames.append(df)

df = pd.concat(frames)
df.head()
```
