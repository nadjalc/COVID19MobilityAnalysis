```python
import numpy as np
import pandas as pd
import matplotlib
import datetime
import time
from itertools import islice
import glob
import io
```

```python
def parse_timestamp(el):
    b = datetime.datetime.strptime(el, fmt)
    return b
```

```python
def parse_timestamp2(el):
    b = datetime.datetime.strptime(el, fmt2)
    return b
```

```python
def create_date_col(df_m):
    date_lst = []
    for index, row in df_m.iterrows():
        day = str(df_m['ride_begin'][index].day)
        month = str(df_m['ride_begin'][index].month)
        year = str(df_m['ride_begin'][index].year)
        date_text = year + '-' + month + '-' + day
        date_lst.append(date_text)
    return date_lst
```

```python
fmt = '%Y-%m-%d'
fmt2 = '%Y-%m-%d %H:%M:%S'
```

```python
li = []
for file in mobility_2020:
    frame = pd.read_csv(file, sep= ';', index_col=None, header=0, low_memory=False)
    li.append(frame) 
df_m20 = pd.concat(li, axis=0, ignore_index=True, sort= False) #connecting all the dates
```

```python
li = []
for file in mobility_2021:
    frame = pd.read_csv(file, sep= ';', index_col=None, header=0, low_memory=False)
    li.append(frame) 
df_m21 = pd.concat(li, axis=0, ignore_index=True, sort= False) #connecting all the dates
```

```python
pd.set_option('display.max_columns()', None)
```

```python
df_c20['date'] = df_c20['date'].map(parse_timestamp)
df_c21['date'] = df_c21['date'].map(parse_timestamp)
df_m20['ride_begin'] = df_m20['ride_begin'].map(parse_timestamp2)
df_m21['ride_begin'] = df_m21['ride_begin'].map(parse_timestamp2)
df_m20['ride_end'] = df_m20['ride_end'].map(parse_timestamp2)
df_m21['ride_end'] = df_m21['ride_end'].map(parse_timestamp2)
```

```python
df_m20['date'] = create_date_col(df_m20)
df_m21['date'] = create_date_col(df_m21)
```

```python
df_m20['date'] = df_m20['date'].map(parse_timestamp)
df_m21['date'] = df_m21['date'].map(parse_timestamp)
```

```python
for index, row in df_c20.iterrows():
    if row['date'].year == 2021:
        df_c20.drop(index, inplace=True)
```

```python
for index, row in df_c21.iterrows():
    if ((row['date'].year == 2021) and (row['date'].month > 7)) or (row['date'].year == 2020):
        df_c21.drop(index, inplace=True)
```

```python
#merging data frames
complete_df20 = pd.merge(df_m20, df_c20, on='date', how='outer')
complete_df21 = pd.merge(df_m21, df_c21, on='date', how='outer')
```

### One day grouping for ride times and distances and further use of mean and median values for correlation analysis, count of rides in a day for correlation analysis

### Correlation of ride times

```python
rt20 = df_m20.groupby('date')['ride_time'].mean()
corr_df20 = pd.merge(df_c20, rt20, on='date', how='left')
```

```python
corr_df20.corr(method ='spearman')
```

```python
rt20 = df_m20.groupby('date')['ride_time'].median()
corr_df20 = pd.merge(df_c20, rt20, on='date', how='left')
```

```python
corr_df20.corr(method ='spearman')
```

```python
rt20 = df_m20.groupby('date')['ride_time'].count()
corr_df20 = pd.merge(df_c20, rt20, on='date', how='left')
```

```python
ab20 = df_m20.groupby('date')['available_bikes'].mean()
corr_df20 = pd.merge(corr_df20, ab20, on='date', how='left')
```
