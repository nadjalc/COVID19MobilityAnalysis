```python
import numpy as np
import pandas as pd
import matplotlib
import datetime
import time
from itertools import islice
```

```python
def parse_timestamp(el):
    b = datetime.datetime.strptime(el, fmt)
    return b
```

```python
def parse_timestamp2(el):
    b = datetime.datetime.strptime(el, fmt2)
    return b
```

```python
fmt = '%Y-%m-%d'
fmt2 = '%Y-%m-%d %H:%M:%S'
```
