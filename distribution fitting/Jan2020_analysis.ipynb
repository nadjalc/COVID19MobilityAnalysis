```python
from scipy.stats import gamma
import matplotlib.pyplot as plt
import numpy as np

def get_statistic(sample, population_cdf):
    emperical = [(1 + i) / len(sample) for i in range(len(sample))]
    real = [population_cdf(s) for s in sample]
    return max([abs(e-r) for e,r in zip(emperical, real)])

shape = 2
scale = 2
population_cdf = lambda x: gamma.cdf(x, shape, scale=scale)
xs = np.arange(0, 20, 0.01)
plt.plot(
    xs,
    [population_cdf(x) for x in xs],
    label='actual distribution'
)

n = 150
sample = np.sort(np.random.gamma(shape, scale, n))
emperical = [(1 + i) / len(sample) for i in range(len(sample))]

statistic = get_statistic(sample, population_cdf)
print('statistic: ', statistic)
plt.step(
    sample,
    emperical,
    label='empirical distribution'
)
plt.legend()
plt.show()
```
