```python
%matplotlib inline

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
import scipy
import scipy.stats as stats
```

```python
matplotlib.rcParams['figure.figsize'] = (12.0, 8.0)
matplotlib.style.use('ggplot')
```

```python
warnings.filterwarnings("ignore")
```

```python
url = "./limited_data_nextbike_2021-01.csv"
df = pd.read_csv(url, sep=';', low_memory=False)
```

### Gamma

```python
ms = df['distance'].to_numpy()
```

```python
scipy.stats.probplot(ms, sparams=(1.37), dist=scipy.stats.gamma, plot=matplotlib.pyplot)
```

### Beta

```python
scipy.stats.probplot(ms, sparams=(1.32, 36), dist=scipy.stats.beta, plot=matplotlib.pyplot)
```
